```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

## Johnson and Johnson Quarterly Earnings (jj)

```python
jj = pd.read_csv('JohnsonJohnson.csv')

jj = jj[['time', 'value']] # ignore the index coliumn

print(jj.shape)

jj.head()
```

The first step in analysis of a time series is basically to produce the time plot

```python
plt.title('Johnson&Johnson quarterly earning per share')
sns.lineplot(x='time', y='value', data=jj)
plt.xlabel('Quarters')
plt.ylabel('Earnings')
plt
plt.show()
```

There some king of trend throughout the years.

We can also see some fluctuations, so there is a seasonal effect in this time series. 

At the beginning of the time series, the variation is not that much, but the leg room i have a higher variations.

this kind of thing violates so-called sationary principle.

## Pneumonia and Influenza deaths in the U.S (flu)

```python
flu = pd.read_csv('flu.csv', names = ['month', 'deaths'], skiprows=1)

print(flu.shape)

flu.head()
```

```python
plt.title('Monthly Pneumonia and Influenza Deaths in US')
sns.lineplot(x='month', y='deaths', data=flu)
plt.xlabel('Months')
plt.ylabel('Number of Deaths per 10.000 people')
plt.show()
```

There is some kind of seasonally going on.

There is a peak every after year or so.

That kind of shows that there is some kind of systally going on in this data which is definitely not a stationary time series

 If we look at the over all trend, there might be a trend that overall number of tasks going down

## Land-ocean Temperature Deviations (globtemp)

```python
globtemp = pd.read_csv('globtemp.csv', names = ['year', 'temp_deviations'], skiprows=1)

print(globtemp.shape)

globtemp.head()
```

```python
plt.title('Global mean land-ocean deviations from average temperature of 1951-1980')
sns.lineplot(x='year', y='temp_deviations', data=globtemp)
plt.xlabel('Years')
plt.ylabel('Temperature Deviations')
plt.show()
```

There's some king of trend. Temperature deviations are going up

Even though there is a trend, there's some king of seasonallity on that trend, probably because of the seasons of the year
