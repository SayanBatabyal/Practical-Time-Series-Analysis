# Moving Average

Linear Combination of the noises that affects it

$$
    X_t = Z_t + \theta_1Z_{t-1}+\theta_2Z_{t-2}
$$

Theta are weights

Moving Average model of order 2: MA(2)

It's called order 2 because we go 2 days back

$$
MA(q)
$$

## Simulating MA(2) Process

We are going to use standard normal distribution for our noise

$$
    X_t = Z_t + 0.7  Z_{t-1}+ 0.2 Z_{t-2}
$$
$$
Z_t ~~ Normal(0,1)
$$

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf

# Generate de Noise
noise = np.random.normal(size=10000)

# Introduce a variable
ma_2 = []

#Loop for generating MA(2) process
for i in range(3,10000):
    ma_2.append(noise[i]+0.7*noise[i-1]+0.2*noise[i-2])
    
# Shift data to left by 2 units
moving_average_process=ma_2[3:10000]

# Put time series structure on vanilla data
ts_moving_average_process = pd.DataFrame(columns=['value'])
#ts_moving_average_process['time'] = pd.to_datetime(range("2014-01-01",2014-01-01) unit="D")
ts_moving_average_process['value'] = moving_average_process

# Partition output graphics as a multi frame of 2 rows and 1 column
plt.figure(figsize=(10,8))
ax1 = plt.subplot(211)
ax2 = plt.subplot(212)

# plot the process and plot its ACF
ts_moving_average_process.plot(ax=ax1, title='A moving average process of order 2')

plot_acf(ts_moving_average_process, title='Correlogram of a moving average process of order 2', lags=40, ax=ax2)

plt.show()
```

I realize there is a high correlation at lag 0. This is always the case

Then there's a high correlation with lag 1.
This make sense because we are still getting noises from the previous step.

There's also noise coming from the 2 steps back, 2 days back, and then nothing.

<b>This is much a characteristic of a MA(q) process.</b><br>
If you are looking at MA(q) process, if you take ACF off MA(q) process, it lags or the autocorreation will cut of at lag q

In this case we have MA(2) process and then ACF has to cut from lag 2
